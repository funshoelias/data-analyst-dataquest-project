# Gunshot in US

## we will investigate gunshot in the us by year, month, race, sex etc

```python
import csv

data_with_header = list(csv.reader(open('guns.csv')))
headers = data_with_header[0]
data = data_with_header[1:]
print(data_with_header[:5])
```

```python
print(headers)
```

```python
print(data[:5])
```

```python
years = [row[1] for row in data]
```

```python
years[:5]
```

```python
year_count = {}

for row in data:
    if row[1] not in year_count:
        year_count[row[1]] = 1
    if row[1] in year_count:
        year_count[row[1]] += 1
```

```python
year_count
```

```python
import datetime

dates = [datetime.datetime(year=int(row[1]), 
                           month=int(row[2]),
                          day=1) for row in data] #day is a fixed day
```

```python
dates[:5]
```

```python
dates_count = {}

for row in dates:
    if row not in dates_count:
        dates_count[row] = 1
    if row in dates_count:
        dates_count[row] += 1
```

```python
dates_count
```

```python
def count(thing_to_count):
    row_count = {}
    for row in data:
        if row[thing_to_count] not in row_count:
            row_count[row[thing_to_count]] = 1
        if row[thing_to_count] in row_count:
            row_count[row[thing_to_count]] += 1
    return row_count
```

```python
sex_counts = count(5)
race_counts = count(7)
```

```python
sex_counts
```

```python
race_counts
```

```python
race_per_hundredk = {}
for key,value in race_counts.items():
    hundredk = (value / mapping[key]) * 100000
    race_per_hundredk[key] = hundredk
    
```
